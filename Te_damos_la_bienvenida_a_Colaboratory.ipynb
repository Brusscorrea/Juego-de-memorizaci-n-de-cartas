```
import random
import time
from IPython.display import clear_output

# Configuración inicial del juego
rows = 2
columns = 3
cards_back = [["|_|" for _ in range(columns)] for _ in range(rows)]
options = ['J', 'J', 'Q', 'Q', 'K', 'K']
game_cards = [["" for _ in range(columns)] for _ in range(rows)]

# Asignar cartas aleatorias a game_cards
for i in range(len(game_cards)):
    for j in range(len(game_cards[i])):
        random_card = random.choice(options)
        game_cards[i][j] = '|' + random_card + '|'
        options.remove(random_card)

# Función para mostrar las cartas
def show_cards(cards):
    for row in cards:
        print(" ".join(row))
        print()

# Definir 't' como el tiempo de espera en segundos
t = 2

# Mostrar todas las cartas al principio para memorizar
show_cards(game_cards)
time.sleep(t)
clear_output(wait=True)

# Juego principal
while '|' + '_' + '|' in str(cards_back):
    show_cards(cards_back)
    time.sleep(t)
    clear_output(wait=True)

    # Primera elección
    row_1 = int(input("Indique la fila para la primera carta: "))
    column_1 = int(input("Indique la columna para la primera carta: "))
    cards_back[row_1][column_1] = game_cards[row_1][column_1]
    show_cards(cards_back)
    time.sleep(t)
    clear_output(wait=True)

    # Segunda elección
    row_2 = int(input("Indique la fila para la segunda carta: "))
    column_2 = int(input("Indique la columna para la segunda carta: "))
    cards_back[row_2][column_2] = game_cards[row_2][column_2]
    show_cards(cards_back)
    time.sleep(t)

    # Comprobar si las cartas coinciden
    if game_cards[row_1][column_1] == game_cards[row_2][column_2]:
        print("¡Encontraste un par!")
        cards_back[row_1][column_1] = "   "
        cards_back[row_2][column_2] = "   "
    else:
        print("Intenta de nuevo.")
        cards_back[row_1][column_1] = "|_|"
        cards_back[row_2][column_2] = "|_|"

    time.sleep(t)
    clear_output(wait=True)

print("¡Felicidades! Has encontrado todos los pares.")
```

CODIGO CON VALIDACION DE ENTRADAS DEL USUARIO Y CONSIDERACION DE CARTAS JUGADAS

```
import random
import time
from IPython.display import clear_output

# Configuración inicial del juego
rows = 2
columns = 3
cards_back = [["|_|" for _ in range(columns)] for _ in range(rows)]
options = ['J', 'J', 'Q', 'Q', 'K', 'K']
game_cards = [["" for _ in range(columns)] for _ in range(rows)]

# Asignar cartas aleatorias a game_cards
for i in range(len(game_cards)):
    for j in range(len(game_cards[i])):
        random_card = random.choice(options)
        game_cards[i][j] = '|' + random_card + '|'
        options.remove(random_card)

# Función para mostrar las cartas
def show_cards(cards):
    for row in cards:
        print(" ".join(row))
        print()

# Definir 't' como el tiempo de espera en segundos
t = 2

# Mostrar todas las cartas al principio para memorizar
show_cards(game_cards)
time.sleep(t)
clear_output(wait=True)

# Función para validar la entrada del usuario
def valid_input(prompt, max_row, max_col):
    while True:
        try:
            value = int(input(prompt))
            if value < 0 or value >= max_row and prompt.lower().startswith("indique la fila") or value >= max_col and prompt.lower().startswith("indique la columna"):
                print(f"Por favor, ingrese un número válido entre 0 y {max_row - 1 if prompt.lower().startswith('indique la fila') else max_col - 1}.")
            else:
                return value
        except ValueError:
            print("Por favor, ingrese un número entero.")

# Juego principal
while '|' + '_' + '|' in str(cards_back):
    show_cards(cards_back)

    # Primera elección
    row_1 = valid_input("Indique la fila para la primera carta: ", rows, columns)
    column_1 = valid_input("Indique la columna para la primera carta: ", rows, columns)
    while cards_back[row_1][column_1] == "   ":
        print("Esa carta ya ha sido jugada. Por favor, elija otra.")
        row_1 = valid_input("Indique la fila para la primera carta: ", rows, columns)
        column_1 = valid_input("Indique la columna para la primera carta: ", rows, columns)
    cards_back[row_1][column_1] = game_cards[row_1][column_1]
    show_cards(cards_back)

    # Segunda elección
    row_2 = valid_input("Indique la fila para la segunda carta: ", rows, columns)
    column_2 = valid_input("Indique la columna para la segunda carta: ", rows, columns)
    while cards_back[row_2][column_2] == "   " or (row_1 == row_2 and column_1 == column_2):
        print("Esa carta ya ha sido jugada o es la misma que la primera elección. Por favor, elija otra.")
        row_2 = valid_input("Indique la fila para la segunda carta: ", rows, columns)
        column_2 = valid_input("Indique la columna para la segunda carta: ", rows, columns)
    cards_back[row_2][column_2] = game_cards[row_2][column_2]
    show_cards(cards_back)

    # Comprobar si las cartas coinciden
    if game_cards[row_1][column_1] == game_cards[row_2][column_2]:
        print("¡Encontraste un par!")
        cards_back[row_1][column_1] = "   "
        cards_back[row_2][column_2] = "   "
    else:
        print("Intenta de nuevo.")
        cards_back[row_1][column_1] = "|_|"
        cards_back[row_2][column_2] = "|_|"

    time.sleep(t)
    clear_output(wait=True)

print("¡Felicidades! Has encontrado todos los pares.")
```
